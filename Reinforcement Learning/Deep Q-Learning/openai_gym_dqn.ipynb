```python
import gym
import torch
import numpy as np
import torch.nn as nn
from tqdm import trange
import torch.optim as optim
from collections import deque
import torch.nn.functional as F
```

```python
class Net(nn.Module):
    '''
    A neural network class with three fully connected layers
    '''
    def __init__(self, lr, input_dims, fc1_dims, fc2_dims, output_dims):
        super(Net, self).__init__()
        
        # Initialise the network's layers
        self.fc1 = nn.Linear(*input_dims, fc1_dims)
        self.fc2 = nn.Linear(fc1_dims, fc2_dims)
        self.fc3 = nn.Linear(fc2_dims, output_dims)
        
        # Initialise the networks parameters
        self.optimiser = optim.Adam(self.parameters(), lr=lr)
        self.loss      = nn.MSELoss()
        self.device    = torch.device('cuda:0')
        self.to(self.device)
        
    def forward(self, x):
        '''
        The forward pass of the network.
        '''
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)
```

```python
class Agent:
    '''
    The Q-Agent Class
    '''
    def __init__(self, lr, discount, batch_size, n_observations, n_actions, \
                 epsilon=1.0, epsilon_dec=5e-4, epsilon_min=0.05, max_mem=100000):
        
        # Initialise network and related parameters
        self.lr             = lr
        self.discount       = discount
        self.batch_size     = batch_size
        self.n_observations = n_observations
        self.n_actions      = n_actions
        self.epsilon        = epsilon
        self.epsilon_dec    = epsilon_dec
        self.epsilon_min    = epsilon_min
        self.mem_size       = max_mem
        self.mem_cntr       = 0
        self.iter_cntr      = 0
        
        # Initialise memory arrays
        self.state_memory     = np.zeros((self.mem_size, *n_observations), dtype=np.float32)
        self.new_state_memory = np.zeros((self.mem_size, *n_observations), dtype=np.float32)
        self.action_memory    = np.zeros(self.mem_size, dtype=np.int32)
        self.reward_memory    = np.zeros(self.mem_size, dtype=np.float32)
        self.terminal_memory  = np.zeros(self.mem_size, dtype=np.bool)
        
        # Initialise an instance of the Net class
        self.Q = Net(lr, n_observations, 256, 256, n_actions)
        
    def store_transition(self, state, action, reward, state_, terminal):
        '''
        Update memory arrays
        '''
        index = self.mem_cntr % self.mem_size
        self.state_memory[index]     = state
        self.new_state_memory[index] = state_
        self.reward_memory[index]    = reward
        self.action_memory[index]    = action
        self.terminal_memory[index]  = terminal

        self.mem_cntr += 1
    
    def choose_action(self, observation):
        '''
        Choose action, either random or through the
        Net class based on the epsilon greedy algorithm.
        '''
        if np.random.random() > self.epsilon:
            # Choose action based on the Net class
            state   = torch.tensor([observation]).to(self.Q.device)
            actions = self.Q.forward(state.type(torch.float32))
            action  = torch.argmax(actions).item()
        else:
            # Choose a random action
            action  = np.random.randint(self.n_actions)
            
        return action
    
    def learn(self):
        if self.mem_cntr < self.batch_size:
            return

        self.Q.optimiser.zero_grad()
        
        max_mem     = min(self.mem_cntr, self.mem_size)
        batch       = np.random.choice(max_mem, self.batch_size, replace=False)
        batch_index = np.arange(self.batch_size, dtype=np.int32)

        # Take a random batch of data
        state_batch     = torch.tensor(self.state_memory[batch]).to(self.Q.device)
        new_state_batch = torch.tensor(self.new_state_memory[batch]).to(self.Q.device)
        action_batch    = self.action_memory[batch]
        reward_batch    = torch.tensor(self.reward_memory[batch]).to(self.Q.device)
        terminal_batch  = torch.tensor(self.terminal_memory[batch]).to(self.Q.device)

        q_eval = self.Q.forward(state_batch)[batch_index, action_batch]# Figure this out
        q_next = self.Q.forward(new_state_batch)
        q_next[terminal_batch] = 0.0 # and this

        q_target = reward_batch + self.discount * torch.max(q_next , dim=1)[0] # understand this more

        # Compute loss and adjust weights accordingly
        loss = self.Q.loss(q_target, q_eval).to(self.Q.device)
        loss.backward()
        self.Q.optimiser.step()

        # Decay epsilon value
        self.epsilon = max(self.epsilon - self.epsilon_dec, self.epsilon_min)
        self.iter_cntr += 1
```

```python
def DQN(env, lr, discount, batch_size, n_episodes):
    env = gym.make(env)
    input_dim = env.observation_space.shape[0]
    output_dim = env.action_space.n
    agent = Agent(lr=lr, discount=discount, batch_size=batch_size,n_observations=[input_dim], n_actions=output_dim)
    scores = []

    for i in (t := trange(n_episodes)):
        score = 0
        done = False
        observation = env.reset()
        while not done:
            action = agent.choose_action(observation)
            observation_, reward, done, info = env.step(action)
            score += reward
            agent.store_transition(observation, action, reward, 
                                    observation_, done)
            agent.learn()
            observation = observation_
        scores.append(score)

        avg_score = np.mean(scores[-100:])
        t.set_description('episode ' + str(i) + ' average score %.2f' % avg_score)
    return scores
```

```python
scores = DQN('CartPole-v1', 0.01, 0.95, 64, 100)
```
